```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.subplot(1, 2, 1) #(row, column, plot number)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2) #plot number 2
plt.plot(y, x, 'b')
```

```python
#Object Oriented Method
```

```python
fig = plt.figure() #This is like a blank canves

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
#Takes in a list (left axis, bottom, width and height relate to the blank canves)

axes.plot(x, y) #This plot
#This is like create an empty canves, add axes, and plot on top of one another.

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #The axes must be zero to 1
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  

axes1.plot(x, y)
axes2.plot(y, x)

axes1.set_title('Larger Plot')
axes2.set_title('Samller Plot')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x, y)
axes[1].plot(y, x)

axes[0].set_title('First Plot')
axes[1].set_title('Second Plot')

plt.tight_layout() #To remove overlapping
```

```python
#Figure Size and DPI
```

```python
fig = plt.figure(figsize = (8, 2), dpi = 100)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(figsize = (8, 2), dpi = 100)

#ax = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8, 2), dpi = 100)

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('my_picture')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label = 'x^2')
ax.plot(x, x**3, label = 'x^3')

#add legend
ax.legend(loc = 0)
```

```python
#Appearance 
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = '#FF8C00', lw = 1, ls = '--', marker = '*', 
        markersize = 20, markerfacecolor = 'blue', alpha = 0.5) 
#RGB Hex Code. Linewidth or lw to change the thickenss. Alpha = transparency
#linestyle or ls and no space. Marker can be o * and anything else. Markersize control its size.
#Markerfacecolor to change its color, markeredgewidth control its edge, and markeredgecolor change the edge color.
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = '#FF8C00', lw = 2, ls = '--')

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
